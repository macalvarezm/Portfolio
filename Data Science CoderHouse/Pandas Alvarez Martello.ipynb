1.Indice con años de los juegos olímpicos

```python
import pandas as pd 
import numpy as np
```

```python
medallas=pd.read_csv('Summer-Olympic-medals-1976-to-2008.csv',encoding= 'unicode_escape')
medallas.head(5)
```

```python
medallas= medallas.reindex(columns=['Year','City','Sport', 'Discipline','Event', 'Athlete', 'Gender', 'Country_Code', 'Country' , 'Event_gender' , 'Medal'])
medallas    
```

```python
indice=list(medallas.Year)
```

```python
columnas=medallas.columns[1:]
columnas
```

```python
valores = medallas.values[:,1:]
```

```python
medallas = pd.DataFrame(valores,index=indice,columns=columnas)
medallas
```

2. Separar nombres de apellidos

```python
Athlete = medallas['Athlete'].str.split('.', n=1, expand=True).rename(columns={0:"First name", 1:"Last name"})
Athlete
```

3. Medidas resumen del conjunto de datos.
   País con mayor cantidad de medallas.

```python
#Cantidad de medallas por país
Med=medallas.groupby('Country')['Medal'].count()
Med
```

```python
#País con mayor cantidad de medallas
Paísmax=medallas.groupby('Country')['Medal'].count().sort_values(ascending=False)[:1]
Paísmax
```

```python
#Medallas en total ganadas por deporte
medallasdeport=medallas.groupby('Sport')['Medal'].count()
medallasdeport
```

```python
#Medallas de oro, plata y bronce por deporte
sportmedal=medallas.groupby('Sport')['Medal'].value_counts()
sportmedal
```

4. Medallas género masculino por año

```python
medalg=pd.read_csv('Summer-Olympic-medals-1976-to-2008.csv',encoding= 'unicode_escape')
medalg= medalg.reindex(columns=['Year','Gender', 'Medal'])
medalg
```

```python
#Filtro por género masculino
medalmen =medalg[medalg['Gender']=='Men']
medalmen
```

```python
menmedal=medalmen.groupby('Year')['Gender'].value_counts()
menmedal
```
